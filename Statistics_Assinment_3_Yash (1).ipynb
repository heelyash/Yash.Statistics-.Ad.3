````python
# Que 1: Q= Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
#   (i) Write a Python function to calculate the mean of a given list of numbers.Create a function to find the median of a list of numbers.
#   (ii) Develop a program to compute the mode of a list of integers.
#   (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
#   (iv) Write a Python function to find the geometric mean of a list of positive numbers.
#   (v) Create a program to calculate the harmonic mean of a list of values.
#   (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
#   (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of  the outlier

import random
from collections import Counter
from scipy import stats
import numpy as np

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]


def calculate_mean(numbers):
  """
  Calculate the mean of a list of numbers.

  Args:
    numbers: A list of numbers.

  Returns:
    The mean of the numbers.
  """
  if not numbers:
    return 0
  return sum(numbers) / len(numbers)


def calculate_median(numbers):
  """
  Calculate the median of a list of numbers.

  Args:
    numbers: A list of numbers.

  Returns:
    The median of the numbers.
  """
  numbers.sort()
  n = len(numbers)
  if n % 2 == 0:
    mid1 = numbers[n // 2 - 1]
    mid2 = numbers[n // 2]
    median = (mid1 + mid2) / 2
  else:
    median = numbers[n // 2]
  return median


def calculate_mode(numbers):
  """
  Calculate the mode of a list of integers.

  Args:
    numbers: A list of integers.

  Returns:
    The mode of the numbers.
  """
  if not numbers:
    return None
  counts = Counter(numbers)
  max_count = max(counts.values())
  mode = [num for num, count in counts.items() if count == max_count]
  return mode


def calculate_weighted_mean(values, weights):
  """
  Calculate the weighted mean of a list of values and their corresponding weights.

  Args:
    values: A list of values.
    weights: A list of weights.

  Returns:
    The weighted mean.
  """
  if not values or not weights:
    return 0
  weighted_sum = sum([val * weight for val, weight in zip(values, weights)])
  total_weight = sum(weights)
  if total_weight == 0:
    return 0
  return weighted_sum / total_weight


def calculate_geometric_mean(numbers):
  """
  Calculate the geometric mean of a list of positive numbers.

  Args:
    numbers: A list of positive numbers.

  Returns:
    The geometric mean.
  """
  if not numbers:
    return 0
  product = 1
  for num in numbers:
    if num <= 0:
      return 0  # Handle non-positive numbers
    product *= num
  return product ** (1 / len(numbers))


def calculate_harmonic_mean(numbers):
  """
  Calculate the harmonic mean of a list of values.

  Args:
    numbers: A list of values.

  Returns:
    The harmonic mean.
  """
  if not numbers:
    return 0
  reciprocals_sum = sum([1 / num for num in numbers if num != 0])
  if reciprocals_sum == 0:
    return 0
  return len(numbers) / reciprocals_sum


def calculate_midrange(numbers):
  """
  Calculate the midrange of a list of numbers.

  Args:
    numbers: A list of numbers.

  Returns:
    The midrange.
  """
  if not numbers:
    return 0
  return (min(numbers) + max(numbers)) / 2


def calculate_trimmed_mean(numbers, percentage):
  """
  Calculate the trimmed mean of a list, excluding a certain percentage of outliers.

  Args:
    numbers: A list of numbers.
    percentage: The percentage of outliers to exclude (e.g., 0.1 for 10%).

  Returns:
    The trimmed mean.
  """
  if not numbers:
    return 0
  numbers.sort()
  n = len(numbers)
  k = int(n * percentage)
  trimmed_numbers = numbers[k:n - k]
  return calculate_mean(trimmed_numbers)


# Example usage:

mean_val = calculate_mean(int_list)
median_val = calculate_median(int_list)
mode_val = calculate_mode(int_list)
weighted_mean_val = calculate_weighted_mean(int_list, [1] * len(int_list))
geometric_mean_val = calculate_geometric_mean(int_list)
harmonic_mean_val = calculate_harmonic_mean(int_list)
midrange_val = calculate_midrange(int_list)
trimmed_mean_val = calculate_trimmed_mean(int_list, 0.1)

print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
print(f"Weighted Mean: {weighted_mean_val}")
print(f"Geometric Mean: {geometric_mean_val}")
print(f"Harmonic Mean: {harmonic_mean_val}")
print(f"Midrange: {midrange_val}")
print(f"Trimmed Mean (10%): {trimmed_mean_val}")
````

````python
# Que 3.: Write a Python class representing a discrete random variable with methods to calculate its expected
# value and variance.

class DiscreteRandomVariable:
  """
  Represents a discrete random variable and provides methods to calculate
  its expected value and variance.
  """

  def __init__(self, probabilities):
    """
    Initializes the discrete random variable with a dictionary of
    probabilities for each possible value.

    Args:
      probabilities: A dictionary where keys are the possible values of the
        random variable, and values are their corresponding probabilities.
    """
    self.probabilities = probabilities

  def expected_value(self):
    """
    Calculates the expected value (mean) of the random variable.

    Returns:
      The expected value.
    """
    expected_value = 0
    for value, probability in self.probabilities.items():
      expected_value += value * probability
    return expected_value

  def variance(self):
    """
    Calculates the variance of the random variable.

    Returns:
      The variance.
    """
    mean = self.expected_value()
    variance = 0
    for value, probability in self.probabilities.items():
      variance += probability * ((value - mean) ** 2)
    return variance


# Example usage:
probabilities = {1: 0.2, 2: 0.3, 3: 0.5}
random_variable = DiscreteRandomVariable(probabilities)

expected_value = random_variable.expected_value()
variance = random_variable.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
````

````python
#Que 4: Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
# variance of the outcomes.

import random

def simulate_die_roll(num_rolls=1000):
  """
  Simulates the rolling of a fair six-sided die and calculates the expected value
  and variance of the outcomes.

  Args:
    num_rolls: The number of times to roll the die.

  Returns:
    A tuple containing the expected value and variance of the outcomes.
  """

  outcomes = [random.randint(1, 6) for _ in range(num_rolls)]
  expected_value = sum(outcomes) / num_rolls
  variance = sum([(x - expected_value)**2 for x in outcomes]) / num_rolls
  return expected_value, variance

# Simulate the die rolls and calculate the expected value and variance
expected_value, variance = simulate_die_roll()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
````

````python
# Que 5.: Create a Python function to generate random samples from a given probability distribution (e.g.,
# binomial, Poisson) and calculate their mean and variance.

def generate_random_samples(distribution, parameters, num_samples):
  """
  Generates random samples from a given probability distribution.

  Args:
    distribution: The name of the distribution (e.g., "binomial", "poisson").
    parameters: A dictionary of parameters for the distribution.
    num_samples: The number of samples to generate.

  Returns:
    A list of random samples.
  """

  if distribution == "binomial":
    n = parameters.get("n", 10)  # Default value for n (number of trials)
    p = parameters.get("p", 0.5)  # Default value for p (probability of success)
    # Use np.random.binomial to generate samples from binomial distribution
    return list(np.random.binomial(n, p, num_samples))
  elif distribution == "poisson":
    lam = parameters.get("lam", 5)  # Default value for lambda (average rate)
    return list(np.random.poisson(lam, num_samples))
  else:
    return []  # Handle unsupported distributions


def calculate_mean_variance(samples):
  """
  Calculates the mean and variance of a list of samples.

  Args:
    samples: A list of numerical samples.

  Returns:
    A tuple containing the mean and variance.
  """

  if not samples:
    return (0, 0)
  mean = np.mean(samples)
  variance = np.var(samples)
  return (mean, variance)


# Example usage
samples_binomial = generate_random_samples("binomial", {"n": 20, "p": 0.6}, 1000)
mean_binomial, variance_binomial = calculate_mean_variance(samples_binomial)
print("Binomial Distribution:")
print("Mean:", mean_binomial)
print("Variance:", variance_binomial)


samples_poisson = generate_random_samples("poisson", {"lam": 3}, 1000)
mean_poisson, variance_poisson = calculate_mean_variance(samples_poisson)
print("\nPoisson Distribution:")
print("Mean:", mean_poisson)
print("Variance:", variance_poisson)
````

````python
# Que 6: Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
# the mean, variance, and standard deviation of the samples

import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
  """
  Generates random samples from a Gaussian (normal) distribution.

  Args:
    mean: The mean of the Gaussian distribution.
    std_dev: The standard deviation of the Gaussian distribution.
    num_samples: The number of samples to generate.

  Returns:
    A list of random samples from the Gaussian distribution.
  """
  return np.random.normal(loc=mean, scale=std_dev, size=num_samples)

def calculate_mean_variance_std(samples):
  """
  Calculates the mean, variance, and standard deviation of a list of samples.

  Args:
    samples: A list of numerical samples.

  Returns:
    A tuple containing the mean, variance, and standard deviation.
  """
  if samples.size == 0:
    return (0, 0, 0)
  mean = np.mean(samples)
  variance = np.var(samples)
  std_dev = np.std(samples)
  return (mean, variance, std_dev)

# Example usage:
mean = 0  # Set the desired mean
std_dev = 1  # Set the desired standard deviation
num_samples = 1000  # Set the number of samples

samples = generate_gaussian_samples(mean, std_dev, num_samples)
mean_calculated, variance_calculated, std_dev_calculated = calculate_mean_variance_std(samples)

print("Generated Samples:")
print(samples)
print("\nCalculated Statistics:")
print("Mean:", mean_calculated)
print("Variance:", variance_calculated)
print("Standard Deviation:", std_dev_calculated)
````

````python
# Que 9:  Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

import seaborn as sns
import numpy as np
import random
from scipy import stats
from collections import Counter
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
from scipy.stats import norm

# ... (Your existing code) ...

def calculate_exponential_cdf(x, lam):
  """
  Calculates the cumulative distribution function (CDF) of an exponential
  distribution.

  Args:
    x: The value at which to evaluate the CDF.
    lam: The rate parameter (lambda) of the exponential distribution.

  Returns:
    The CDF value at x.
  """
  if x < 0:
    return 0
  else:
    return 1 - np.exp(-lam * x)

# Example usage:
lam = 0.5  # Rate parameter (lambda)
x = 2  # Value at which to evaluate the CDF

cdf_value = calculate_exponential_cdf(x, lam)
print(f"The CDF value at x={x} for an exponential distribution with lambda={lam} is: {cdf_value}")


# Optional: Visualize the CDF
x_values = np.linspace(0, 10, 100)  # Create a range of x values
cdf_values = [calculate_exponential_cdf(x, lam) for x in x_values]

plt.plot(x_values, cdf_values)
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.title('CDF of Exponential Distribution (lambda = {})'.format(lam))
plt.grid(True)
plt.show()
````

````python
# Que 10: ) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

import math

def poisson_pmf(k, lam):
  """
  Calculates the probability mass function (PMF) of a Poisson distribution.

  Args:
    k: The number of events.
    lam: The average rate of events (lambda).

  Returns:
    The PMF value at k.
  """
  if k < 0:
    return 0
  else:
    return (math.exp(-lam) * (lam**k)) / math.factorial(k)

# Example usage:
lam = 3  # Average rate of events
k = 2  # Number of events

pmf_value = poisson_pmf(k, lam)
print(f"The PMF value at k={k} for a Poisson distribution with lambda={lam} is: {pmf_value}")
````

````python
#Que  11. : ;;M A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
# who make a purchase). They collect data from the old and new layouts to compare.
# To generate the data use the following command:
# ```python
# import numpy as np
# # 50 purchases out of 1000 visitors
# old_layout = np.array([1] * 50 + [0] * 950)
# # 70 purchases out of 1000 visitors
# new_layout = np.array([1] * 70 + [0] * 930)
#   ```
# Apply z-test to find which layout is successful.

import numpy as np
from scipy.stats import norm

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate conversion rates
old_conversion_rate = np.mean(old_layout)
new_conversion_rate = np.mean(new_layout)

# Perform a two-sample z-test for proportions
n1 = len(old_layout)
n2 = len(new_layout)
p1 = old_conversion_rate
p2 = new_conversion_rate

# Calculate the pooled proportion
pooled_p = (np.sum(old_layout) + np.sum(new_layout)) / (n1 + n2)

# Calculate the standard error
se = np.sqrt(pooled_p * (1 - pooled_p) * (1 / n1 + 1 / n2))

# Calculate the z-score
z_score = (p2 - p1) / se

# Calculate the p-value
p_value = norm.sf(z_score)

print(f"Old Conversion Rate: {old_conversion_rate:.4f}")
print(f"New Conversion Rate: {new_conversion_rate:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("The new website layout leads to a statistically significant higher conversion rate.")
else:
    print("There is no statistically significant difference in conversion rates between the old and new layouts.")
````

````python
# Que 12 : ;M A tutoring service claims that its program improves students' exam scores. A sample of students who
# participated in the program was taken, and their scores before and after the program were recorded.
# Use the below code to generate samples of respective arrays of marks:
# ```python
# before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
# after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])
# ```
# Use z-test to find if the claims made by tutor are true or fal

import numpy as np
from scipy.stats import ttest_ind

# Sample data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Perform an independent samples t-test
t_statistic, p_value = ttest_ind(before_program, after_program)

print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
  print("The tutoring program leads to a statistically significant improvement in exam scores.")
else:
  print("There is no statistically significant difference in exam scores before and after the program.")
````

````python
# Que 14: 29U A customer service department claims that their average response time is less than 5 minutes. A sample
# of recent customer interactions was taken, and the response times were recorded.
# Implement the below code to generate the array of response time:
# ```python
# response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])
# ```
# Implement z-test to find the claims made by customer service department are tru or false.

import numpy as np
from scipy.stats import norm

# Generate the array of response times
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Population mean claimed by the department
population_mean = 5

# Sample mean
sample_mean = np.mean(response_times)

# Sample standard deviation
sample_std = np.std(response_times)

# Sample size
n = len(response_times)

# Calculate the z-score
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Calculate the p-value (one-tailed test, as the claim is about "less than")
p_value = norm.cdf(z_score)

# Significance level (alpha)
alpha = 0.05

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

if p_value < alpha:
  print("The claim that the average response time is less than 5 minutes is supported by the data.")
else:
  print("The claim that the average response time is less than 5 minutes is not supported by the data.")
````

````python
# Que 15: 2U A company is testing two different website layouts to see which one leads to higher click-through rates.
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
# freedom, and p-value.
# Use the following data:
# ```python
# layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
# layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

import numpy as np
from scipy.stats import ttest_ind

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
  """
  Performs an A/B test analysis to compare click-through rates between two layouts.

  Args:
    layout_a_clicks: A list of click counts for layout A.
    layout_b_clicks: A list of click counts for layout B.

  Returns:
    A tuple containing the t-statistic, degrees of freedom, and p-value.
  """

  t_statistic, p_value = ttest_ind(layout_a_clicks, layout_b_clicks)
  degrees_of_freedom = len(layout_a_clicks) + len(layout_b_clicks) - 2

  return t_statistic, degrees_of_freedom, p_value


# Example usage with provided data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_statistic, degrees_of_freedom, p_value = ab_test_analysis(
    layout_a_clicks, layout_b_clicks
)

print("T-statistic:", t_statistic)
print("Degrees of freedom:", degrees_of_freedom)
print("P-value:", p_value)
````

````python
# Que 17: 2.U A school district introduces an educational intervention program to improve math scores. Write a Python
# function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
# determine if the intervention had a significant impact.
# Use the following data of test score:
#   ```python
#   pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
#   post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

from scipy.stats import ttest_ind

def analyze_intervention(pre_intervention_scores, post_intervention_scores):
  """
  Analyzes pre- and post-intervention test scores to determine if the intervention
  had a significant impact on math scores.

  Args:
    pre_intervention_scores: A list of pre-intervention test scores.
    post_intervention_scores: A list of post-intervention test scores.

  Returns:
    A tuple containing the t-statistic and p-value.
  """

  t_statistic, p_value = ttest_ind(pre_intervention_scores, post_intervention_scores)
  return t_statistic, p_value


# Example usage with the provided data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

t_statistic, p_value = analyze_intervention(
    pre_intervention_scores, post_intervention_scores
)

print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
  print("The educational intervention had a statistically significant impact on math scores.")
else:
  print("There is no statistically significant difference in math scores before and after the intervention.")
````
